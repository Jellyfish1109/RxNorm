```python
import pandas as pd
import requests
import time 
from pathlib import Path
```

```python
input_unique_atc = Path("/Users/theadiselab/Desktop/Jun Hyuk Lee_files/Files/unique_atc_classification.xlsx")

unique_df = pd.read_excel(input_unique_atc, sheet_name = "unique_ATC_classes")
```

```python
def fetch_atc_code_by_name(class_name: str) -> str: 
    url = "https://rxnav.nlm.nih.gov/REST/rxclass/class/byName.json"
    params = {
        "className": class_name,
        "relaSource": "ATC"
    }

    if API_KEY:
        params["apiKey"] = API_KEY

    try: 
        resp = requests.get(url, params = params, timeout = 5)
        resp.raise_for_status()
        data = resp.json()

        min_list = data.get("rxclassMinConceptList", {}).get("rxclassMinConcept", [])
        if isinstance(min_list, list) and min_list:
            first_item = min_list[0]
            return first_item.get("classId", "")
        else: 
            return ""
    except Exception as e: 
        return ""
```

```python
unique_df["atc_code"] = ""

for idx, row in unique_df.iterrows():
    class_name = str(row["unique_atc_classification"]).strip()
    if not class_name: 
        continue
    
    code = fetch_atc_code_by_name(class_name)
    unique_df.at[idx, "atc_code"] = code

    time.sleep(0.1)

    print("Unique ATC code fetched")
```

```python
output_path = Path("/Users/theadiselab/Desktop/Jun Hyuk Lee_files/Files/unique_atc_code.xlsx")
with pd.ExcelWriter(output_path, engine="xlsxwriter") as writer:
    unique_df.to_excel(writer, index=False, sheet_name="unique_atc_code")

print(f"Done!")
```
